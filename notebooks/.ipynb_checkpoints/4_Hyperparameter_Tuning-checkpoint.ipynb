# Module Imports

```python
import pandas as pd
import os
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import applications, layers, callbacks
import keras_tuner as kt
from ast import literal_eval
from skmultilearn.model_selection import iterative_train_test_split
import numpy as np
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
```

# Hyperparameters
Hyper parameters are set at the start of this notebook that will be used throughout.

```python
IMG_DIM = 299
IMG_SIZE = (IMG_DIM, IMG_DIM)
BATCH_SIZE=16
TEST_VAL_SIZE = 0.2
AUTOTUNE = tf.data.AUTOTUNE
EPOCHS = 100
BASE_LR = 0.001
PATIENCE = 10
SEED = 100
```

# Load Dataset

```python
def dataset_from_dataframe(dataframe, class_names):
    img = dataframe.filename
    labels = dataframe[class_names]
    return tf.data.Dataset.from_tensor_slices((img, labels))
   
def decode_img(filename, img_size):
    img_dir = "../data/images/"
    filepath = img_dir + filename
    img = tf.io.read_file(filepath)
    img = tf.io.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, img_size)
    return img

def process_dataset(filename, labels, img_size=IMG_SIZE, return_filenames=False):
    img = decode_img(filename, img_size=img_size)
    if return_filenames:
        return img, labels, filename
    else:
        return img, labels

def configure_dataset_for_performance(dataset, shuffle=False, batch_size=BATCH_SIZE):
    dataset = dataset.cache()
    if shuffle:
        dataset = dataset.shuffle(buffer_size=dataset.cardinality(), reshuffle_each_iteration=True)
    dataset = dataset.batch(batch_size=batch_size, num_parallel_calls=AUTOTUNE)
    dataset = dataset.prefetch(buffer_size=AUTOTUNE)
    return dataset
```

### Dataframes Splits
The filtered dataframe is loaded and split using stratification.

```python
# Read filtered csv file into dataframe
csv_path = "../data/csv_files/classes_filtered.csv"
df = pd.read_csv(csv_path)
# Change diseaes col to be lists. Create class_names and num_classes variables
df.Diseases = df.Diseases.apply(lambda x: literal_eval(x))
class_names = df.columns[1:-2]
num_classes = len(class_names)

# Create X and y arrays, of filenames and class binary coding columns 
X = df.filename.to_frame().to_numpy()
y = df[class_names].to_numpy()

# Iterative train test split. Predifined proportion declared in hyperparameters above.
X_train, y_train, X_test, y_test = iterative_train_test_split(X=X, y=y, test_size=TEST_VAL_SIZE)
# Create dataframes for all splits
X_train = pd.DataFrame(X_train, columns = ["filename"])
y_train = pd.DataFrame(y_train, columns = class_names)
X_test = pd.DataFrame(X_test, columns = ["filename"])
y_test = pd.DataFrame(y_test, columns = class_names)
# Merge dataframes into train and test
train_df = X_train.merge(y_train, left_index=True, right_index=True)
test_df = X_test.merge(y_test, left_index=True, right_index=True)

# Repeat above process, splitting the test set into validation and test, 50/50 split.
X = test_df.filename.to_frame().to_numpy()
y = test_df[class_names].to_numpy()

X_val, y_val, X_test, y_test = iterative_train_test_split(X=X, y=y, test_size=0.5)
X_val = pd.DataFrame(X_val, columns = ["filename"])
y_val = pd.DataFrame(y_val, columns = class_names)
X_test = pd.DataFrame(X_test, columns = ["filename"])
y_test = pd.DataFrame(y_test, columns = class_names)

val_df = X_val.merge(y_val, left_index=True, right_index=True)
test_df = X_test.merge(y_test, left_index=True, right_index=True)
```

```python
print(f"CSV path: {csv_path}")
print(f"Dataframe Shape: {df.shape}")
print(f"Class Names: {list(class_names)}")
print(f"Num Classes: {num_classes}")
print("Dataset shapes:")
print(f"Train DF: {train_df.shape}")
print(f"Val DF: {val_df.shape}")
print(f"Test DF: {test_df.shape}")
total_samples = len(train_df) + len(val_df) + len(test_df)
print(f"Total Samples: {total_samples}")
display(df.head(3))
```

### Dataset
A Tf.data.dataset is created from each of the split dataframes.

```python
# Create datasets which returns filename and binary encoded labels
train_ds = dataset_from_dataframe(train_df, class_names).shuffle(len(train_df))
val_ds = dataset_from_dataframe(val_df, class_names).shuffle(len(val_df))
test_ds = dataset_from_dataframe(test_df, class_names).shuffle(len(test_df))

# Convert filename into image pixels array
train_ds1 = train_ds.map(lambda x,y:process_dataset(x, y, return_filenames=False))
val_ds1  = val_ds.map(lambda x,y:process_dataset(x, y, return_filenames=False))
test_ds1 = test_ds.map(lambda x,y:process_dataset(x, y, return_filenames=False))

print(f"Total no.: samples: {len(df)}")
print(f"Training Samples: {len(train_ds1)}")
print(f"Validation Samples: {len(val_ds1)}")
print(f"Testing Samples: {len(test_ds1)}")
print()
for i in train_ds.element_spec:
    print(i)

for x, y in train_ds.take(1):
    print(x)
    print(y)

# Optimize tf dataset
train_ds2 = configure_dataset_for_performance(train_ds1, shuffle=True)
val_ds2 = configure_dataset_for_performance(val_ds1)
test_ds2 = configure_dataset_for_performance(test_ds1)
```

# Hyperparameter Tuning
- The code for the model is defined in "call_existing_code".
- The hyperparameter options are specified in "build_model"

```python
def call_existing_code(units, dropout_rate, activation_fn):
    base_model = applications.MobileNetV2(include_top=False, input_shape=(224,224,3))
    base_model.trainable = False

    inputs = keras.Input(shape=base_model.input_shape[1:])
    x = applications.mobilenet_v2.preprocess_input(inputs)
    x = base_model(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(dropout_rate)(x)
    
    if units != 0:
        x = layers.Dense(units=units, use_bias=False)(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
        
    outputs = layers.Dense(num_classes, activation=activation_fn, name="clasifier")(x)
    model = keras.Model(inputs, outputs)

    binary = True if activation_fn == "sigmoid" else False

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.0001),
        loss="binary_crossentropy" if binary else "categorical_crossentropy",
        metrics=["binary_accuracy" if binary else "accuracy",
                 keras.metrics.F1Score(average="weighted",
                                       threshold=0.5 if binary else None,
                                       name="f1_score_weighted", dtype=None),
                 keras.metrics.F1Score(average=None,
                                       threshold=0.5 if binary else None,
                                       name="f1_score_per_class", dtype=None)]
    )
    return model

def build_model(hp):
    units = hp.Choice("units", [0, 256, 512])
    dropout_rate = hp.Choice("dropout", [0.2, 0.5])
    activation_fn = hp.Choice("activation_fn", ["sigmoid", "softmax"])

    model = call_existing_code(
        units=units, dropout_rate=dropout_rate, activation_fn=activation_fn
    )
    print(model.get_compile_config())
    return model

def apply_resize(x, img_dim):
    return layers.Resizing(img_dim, img_dim)(x)
```

### Tuner
- The RandomSearch tuner is used to identify the best performing hyperparameters.
- overwrite is set to False, in the event of system crashing, the training can be resumed from the saved logs.
- A random seed is used for replication.
- The tuner is searching to identify the paraeters that yield the lowest validation loss.

```python
tuner = kt.RandomSearch(
    hypermodel= build_model,
    objective = kt.Objective("val_loss", "min"),
    directory="../hyperparameterTuning/",
    project_name="mobilenet_HT",
    overwrite=False,
    max_trials=40,
    seed=1
)
tuner.search_space_summary()
```

# Preparation and Searching
The data is prepared, and the tuner searches the search space.

```python
img_dim = 224
train_ds = train_ds2.map(lambda x,y: (apply_resize(x, img_dim), y), num_parallel_calls=AUTOTUNE)
val_ds = val_ds2.map(lambda x,y: (apply_resize(x, img_dim), y), num_parallel_calls=AUTOTUNE)
test_ds = test_ds2.map(lambda x,y: (apply_resize(x, img_dim), y), num_parallel_calls=AUTOTUNE)
```

```python
cbs = [callbacks.EarlyStopping(patience=PATIENCE, restore_best_weights=True, verbose=1)]

tuner.search(train_ds, validation_data=val_ds, epochs=EPOCHS, callbacks=cbs)
```

```python
tuner.results_summary(32)
```

# Get Best Epochs
- After identifying the best hyperpparameters, the model is trained again to dentify the best epoch, as determined by the lowest validation loss.
- Only the top classifier is trained.

```python
best_hps = tuner.get_best_hyperparameters(1)[0]
print(best_hps.values)
model = tuner.hypermodel.build(best_hps)

model.summary(layer_range=(model.get_layer(index=1).name, model.get_layer(index=-1).name ), expand_nested=True, show_trainable=True)
```

```python
cbs = [callbacks.EarlyStopping(monitor="val_loss", patience=PATIENCE, restore_best_weights=True, verbose=1)]

history_clf = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=cbs
)
```

```python
history_clf_df = pd.DataFrame(history_clf.history)
history_clf_df["epoch"] = history_clf.epoch
best_clf_epoch = history_clf_df[history_clf_df.val_loss == history_clf_df.val_loss.min()]
best_clf_epoch_num = best_clf_epoch.epoch.values[0] + 1
best_clf_epoch_num
```

```python
axes = history_clf_df.plot(
    y=["loss", "val_loss", "binary_accuracy", "val_binary_accuracy", "f1_score_weighted", "val_f1_score_weighted"],
    layout=(1,3),
    subplots=[("loss", "val_loss"), ("binary_accuracy", "val_binary_accuracy"), ("f1_score_weighted", "val_f1_score_weighted")],
    figsize=(15, 5),
    grid=True,
    colormap=ListedColormap(["tab:blue", "tab:orange"]*3),
    style=["-","--"]*3,
    title=["Loss", "Binary_accuracy", "F1_score"]*2
);
for ax in axes.flatten():
    ax.axvline(best_clf_epoch_num-1, color="green", linestyle=":", label="best_epoch")
    ax.legend()
```

# Retrain Best Model
- The model is then finally trained using the best hyperparameters and for the best epoch number.
- The training and validation datasets are combined so as to provide the model with more data during training. (https://keras.io/guides/keras_tuner/getting_started/#tune-endtoend-workflows)

```python
train_ds_all = train_ds1.concatenate(val_ds1)
train_ds_all = configure_dataset_for_performance(train_ds_all, shuffle=True)

img_dim = 224
train_ds_all = train_ds_all.map(lambda x,y: (apply_resize(x, img_dim), y), num_parallel_calls=AUTOTUNE)
test_ds = test_ds2.map(lambda x,y: (apply_resize(x, img_dim), y), num_parallel_calls=AUTOTUNE)
```

```python
model = tuner.hypermodel.build(best_hps)
model.summary(layer_range=(model.get_layer(index=1).name, model.get_layer(index=-1).name ), expand_nested=True, show_trainable=True)
```

```python
history_best_clf = model.fit(
    train_ds_all,
    epochs=best_clf_epoch_num,
)
```

```python
history_best_clf_df = pd.DataFrame(history_best_clf.history)
history_best_clf_df.head()
```

### Evaluate Test Set
- The test results are returned by evaluate the model on the test set.

- The results on the train and test sets are displayed, and a graph also displays the training history.

```python
test_metrics_dict = {"loss":test_loss, "binary_accuracy":test_binary_accuracy, "f1_score_weighted":test_f1_score_per_weighted}
test_metrics_df = pd.DataFrame(test_metrics_dict, index=[0])
test_metrics_df["dataset"] = "test"
test_metrics_df.at[0, "f1_score_per_class"] = str(test_f1_score_per_class.numpy().tolist())
test_metrics_df.to_csv("../hyperparameterTuning/mobilenet_HT_TEST_history.csv", index=False)
```

```python
train_last_epoch = history_best_clf_df.tail(1).copy()
train_last_epoch.loc[:, "dataset"] = "train"

test_vs_train = pd.concat([test_metrics_df, train_last_epoch], join="inner").reset_index(drop=True)
test_vs_train.index = test_vs_train["dataset"]
test_vs_train = test_vs_train.drop(columns=["dataset"]).sort_values(by="loss", ascending=True)

test_vs_train2 = test_vs_train.copy()
test_vs_train2.binary_accuracy = np.round(test_vs_train2.binary_accuracy * 100, 2).astype(str) + "%"
test_vs_train2.f1_score_weighted = np.round(test_vs_train2.f1_score_weighted * 100, 2).astype(str) + "%"

test_vs_train2 = np.round(test_vs_train2, 4)
test_vs_train2 = test_vs_train2.drop(columns=["f1_score_per_class"])
test_vs_train2
```

```python
cols = ["loss", "binary_accuracy", "f1_score_weighted"]
axes = history_best_clf_df[cols].plot(
    y = cols,
    layout=(1,3),
    subplots=True,
    grid=True,
    figsize=(15,5),
    title=list(test_vs_train2.columns),
    xlabel="epochs"
);
for i, ax in enumerate(axes.flatten()):
    metric_name = cols[i]
    metric_val = test_metrics_dict[metric_name]
    ax.axhline(metric_val, color="red", linestyle="--", label="test_value")
    ax.legend()
    ax.plot(history_best_clf_df.index.max(), metric_val, marker="X", c="r", markersize=10)
```
